```python
import geopandas as gpd

# Shapefile de comunas
comunas = gpd.read_file("https://github.com/AdriMA3/introgeodf/raw/refs/heads/main/maps/Comunas/comunas.shp")

# Verifica el nombre de la columna de comuna
comunas.columns  # Busca algo como 'NOM_COM' o 'Comuna'
```

```python
# Filtrar Santiago
santiago = comunas[comunas['Comuna'].str.contains('Santiago', case=False)]
santiago
```

```python
from libpysal.weights import Rook, Queen, KNN

# Convertir a índice continuo
comunas = comunas.reset_index(drop=True)

# Crear objetos de vecindad
w_rook = Rook.from_dataframe(comunas)
w_queen = Queen.from_dataframe(comunas)
w_knn8 = KNN.from_dataframe(comunas, k=8)
```

```python
idx_santiago = comunas[comunas['Comuna'].str.contains('Santiago', case=False)].index[0]
```

```python
base = comunas.iloc[w_rook.neighbors[idx_santiago]].plot(facecolor="yellow", edgecolor="black")
comunas.loc[[idx_santiago]].plot(ax=base, facecolor="red")
```

```python
base = comunas.iloc[w_queen.neighbors[idx_santiago]].plot(facecolor="yellow", edgecolor="black")
comunas.loc[[idx_santiago]].plot(ax=base, facecolor="red")
```

```python
base = comunas.iloc[w_knn8.neighbors[idx_santiago]].plot(facecolor="yellow", edgecolor="black")
comunas.loc[[idx_santiago]].plot(ax=base, facecolor="red")
```

```python
import numpy as np
comunas_final.loc[:, 'dat_alfabetismo'] = comunas_final['dat_alfabetismo'].replace(' -',np.nan)
```

```python
# Revisar columnas
comunas_final.columns

# Elige la variable que quieras (asegúrate que no tenga NaNs)
comunas_final['dat_alfabetismo'] = comunas_final['dat_alfabetismo'].astype(float)
comunas_final = comunas_final.dropna(subset=['dat_alfabetismo'])
```

```python
from esda.moran import Moran

w_knn8 = KNN.from_dataframe(comunas_final, k=8)
w_knn8.transform = 'R'

moran_alf = Moran(comunas_final['dat_alfabetismo'], w_knn8)

print("Moran's I:", moran_alf.I)
print("p-value:", moran_alf.p_sim)
```

```python
from splot.esda import moran_scatterplot
import matplotlib.pyplot as plt

fig, ax = moran_scatterplot(moran_alf, aspect_equal=True)
ax.set_title("Moran Scatterplot - Alfabetismo")
ax.set_xlabel("Alfabetismo (estandarizado)")
ax.set_ylabel("Lag Espacial de Alfabetismo")
plt.show()
```

```python
from esda.moran import Moran_Local

lisa_disc = Moran_Local(y=comunas_final['dat_discapacidad'], w=w_knn8, seed=2022)
```

```python
# Asignar etiquetas solo si p < 0.05
comunas_final['disc_quadrant'] = [l if p < 0.05 else 0 for l, p in zip(lisa_disc.q, lisa_disc.p_sim)]

# Mapear nombres
labels = ['0 no_sig', '1 hotSpot', '2 coldOutlier', '3 coldSpot', '4 hotOutlier']
comunas_final['disc_quadrant_name'] = [labels[i] for i in comunas_final['disc_quadrant']]
```

```python
from libpysal.weights import KNN

# Crear vecinos KNN con CRS en metros
w_knn8_validas = KNN.from_dataframe(comunas_validas, k=8)
w_knn8_validas.transform = 'R'
```

Vemos cual es mejor el más cercano a 1

```python
from esda import shape as shapestats

resultados_shape = {}

for tipo in ["hc_ag6", "hc_ag6_knn8"]:
    regiones = comunas_validas[[tipo, "geometry"]].to_crs(5361).dissolve(by=tipo)
    ipq = shapestats.isoperimetric_quotient(regiones)
    resultados_shape[tipo] = ipq

pd.DataFrame(resultados_shape)
```

Vemos cual de los dos tiene mejor calidad de clustering

```python
from sklearn import metrics

fit_scores = []

for cluster_type in ["hc_ag6", "hc_ag6_knn8"]:
    ch_score = metrics.calinski_harabasz_score(
        comunas_validas[selected_variables_new_std],
        comunas_validas[cluster_type]
    )
    sil_score = metrics.silhouette_score(
        comunas_validas[selected_variables_new_std],
        comunas_validas[cluster_type]
    )
    fit_scores.append((cluster_type, ch_score, sil_score))

pd.DataFrame(fit_scores, columns=["cluster type", "CH score", "SIL score"]).set_index("cluster type")
```
